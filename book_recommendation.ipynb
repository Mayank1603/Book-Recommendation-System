#POPULARITY BASED

```python
ratings_with_name = ratings.merge(books, on='ISBN')
ratings_with_name
```

```python
num_rating = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
num_rating
```

```python
popular = num_rating
popular
```

```python
popular[popular['num_ratings']>=250].sort_values('num_ratings', ascending=False).head(50)
popular.merge(books, on='Book-Title').drop_duplicates('Book-Title')[['Book-Title', 'Book-Author', 'Image-URL-M', 'num_ratings']]
popular
```

```python
x = ratings_with_name.groupby("User-ID").count()['Book-Rating']>200
p = x[x].index
p
```

```python
filtered = ratings_with_name[ratings_with_name['User-ID'].isin(p)]
y = filtered.groupby('Book-Title').count()['Book-Rating']>=50
famous = y[y].index
final_rating = filtered[filtered['Book-Title'].isin(famous)]
final_rating
```

```python
pt = final_rating.pivot_table(index='Book-Title', columns='User-ID', values='Book-Rating')
pt.fillna(0, inplace=True)
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity_scores = cosine_similarity(pt)
similarity_scores.shape
```

```python
def recommend(book_name):
    #index fetch
    index = np.where(pt.index == book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])), key=lambda x:x[1], reverse=True)[1:6]
    for i in similar_items:
        print(pt.index[i[0]])
```

```python
recommend("1st to Die: A Novel")
```
